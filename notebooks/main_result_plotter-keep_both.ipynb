```python
comb_cm_effects, new_names = combine_npi_samples(main_panel_dict, np.array(res_yaml["alpha_i"]), res_yaml["cm_names"])
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Optima"
plt.rcParams["font.weight"] = "light"
    
npi_cols = [d['color'] for k, d in main_panel_dict.items()]
    
plt.figure(figsize=(4, 4), dpi=300)
per_red = 100 * (1 - np.exp(-comb_cm_effects))
median_alpha = np.median(per_red, axis=0)
nS, nCMs = per_red.shape

for i in range(0, nCMs, 2):
    plt.fill_between(
        [-100, 100],
        [-i + 0.5, -i + 0.5],
        [-i - 0.5, -i - 0.5],
        color="tab:grey",
        alpha=0.1,
        linewidth=0,
    )

li, lq, uq, ui = np.percentile(per_red, [2.5, 25, 75, 97.5], axis=0)

for n in range(nCMs):
    plt.plot([li[n], ui[n]], [-n, -n], color=npi_cols[n], alpha=0.1)
    plt.plot([lq[n], uq[n]], [-n, -n], color=npi_cols[n], alpha=0.5)
                         
plt.plot([0, 0], [0.5, -nCMs - 2], "k--", linewidth=0.5)

plt.yticks(-np.arange(len(median_alpha)), new_names)
    
for i, tick in enumerate(plt.gca().get_yticklabels()):
    tick.set_color(npi_cols[i])
    
plt.scatter(median_alpha, -np.arange(len(median_alpha)), marker="|", color=npi_cols)
plt.xlabel("Reduction in R (%)")
plt.ylim([-nCMs + 0.5, 0.5])
plt.xlim([-10, 50])

plt.title("Keep Both")
```
